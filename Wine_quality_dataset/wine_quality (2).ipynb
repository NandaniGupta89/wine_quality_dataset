```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(12, 10))

axes = axes.flatten()

for i, column in enumerate(data.columns):

    sns.histplot(data[column], ax=axes[i], kde=True, color='skyblue')
    axes[i].set_title(f'Histogram and KDE for {column}')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(16, 12))

axes = axes.flatten()


for i, column in enumerate(data.columns):

    sns.boxplot(x=data[column], ax=axes[i], color='skyblue')
    axes[i].set_title(f'Box Plot for {column}')

plt.tight_layout()
plt.show()
```

```python
frequency_table = data['quality'].value_counts().sort_index().reset_index()
frequency_table.columns = ['Quality', 'Frequency']
print(frequency_table)
plt.figure(figsize=(8, 6))
sns.barplot(x='Quality', y='Frequency', data=frequency_table, palette='viridis')
plt.title('Frequency Distribution of Quality')
plt.xlabel('Quality')
plt.ylabel('Frequency')
plt.show()
```

```python
correlation_matrix = data.corr()


plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()


highly_correlated_pairs = []

for i in range(len(correlation_matrix.columns)):
  for j in range(i):
    if abs(correlation_matrix.iloc[i, j])>highly_correlated_pairs:
      pair = (correlation_matrix.columns[i], correlation_matrix.columns[j])
      highly_correlated_pairs.append(pair)

print("Variables that are highly correlated:")
for pair in highly_correlated_pairs:
    print(f"{pair[0]} and {pair[1]} with correlation {correlation_matrix.loc[pair[0], pair[1]]:.2f}")
```

```python
correlation_matrix = data.corr()


plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()


threshold = 0.7  # Adjust the threshold as needed
highly_correlated_pairs = []

for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            pair = (correlation_matrix.columns[i], correlation_matrix.columns[j])
            highly_correlated_pairs.append(pair)

print("Variables that are highly correlated:")
for pair in highly_correlated_pairs:
    print(f"{pair[0]} and {pair[1]} with correlation {correlation_matrix.loc[pair[0], pair[1]]:.2f}")
```
